# Goal 7 data exploration
This workbook aims to analyse all the series available for goal 7, one of the SDGs of the year for the 2023 cycle. To give a better view of the progress on the SDGs in the region level ([M49 standards](https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=&cad=rja&uact=8&ved=2ahUKEwjC9ITkhdr7AhVeM1kFHYv3BjIQFnoECBcQAQ&url=https%3A%2F%2Funstats.un.org%2Funsd%2Fmethodology%2Fm49%2F&usg=AOvVaw2g_oo_TTu8bARRq5C3lCZ2)). At the end of the workbook, an ad hoc analysis was made to understand the disggregation by urban/rural across the UN regions.

This python notebook relies on 4 custom-defined functions from the "*int_func.py*" script:
*  ` return_seriesCode`  
Function to return the SDG series code given the indicator(s) 
*  `return_datapoints`  
Function to return a timeseries data given the geoArea code and the SDG series
*  `return_annualpoints`  
Function to return a cross-sectional snapshot of the global performance on a given SDG series
*  `aggregate_val`  
Function to return a dataframe with disaggregation dimension to only the aggregate
*  `regional_analysis_vis`  
Function plotting the time series data for 8 UN regions, including the *World* (001). Where data is not available, the grid plot shows only the the line plot of other regions in grey.

### Miscellaneous
Other ad hoc detail analysis on SDG series.

```python
full_df = pd.DataFrame()
for key, value in region_dict.items():
    print(value + " downloading")
    df = return_datapoints('EG_ACS_ELEC', disagg = True, geoAreaCode=key, plot = False)
    df['REGION'] = [value for x in range(len(df))]
    df.reset_index(inplace = True)
    full_df = full_df.append(df)
```

```python
full_df
```

```python
allarea = full_df.pivot(index = 'timePeriodStart',columns='REGION',values = 'ALLAREA')
allarea.plot()
```

```python
urban = full_df.pivot(index = 'timePeriodStart',columns='REGION',values = 'URBAN')
urban.plot()
```

```python
rural = full_df.pivot(index = 'timePeriodStart',columns='REGION',values = 'RURAL')
rural.plot()
```
